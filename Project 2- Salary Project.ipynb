```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv(r"C:\Users\s0116731\Desktop\salaries_dataset.csv")
salary = data.drop(['rank','discipline','sex','yrs.since.phd'],axis=1)
print(salary)
```

```python
from sklearn.model_selection import train_test_split
X = salary[['yrs.service']]
y = salary['salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
salary = scale.fit_transform(salary)

from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

y_predict = linear_reg.predict(X_test)
pd.DataFrame({'Predicted': y_predict})
```
